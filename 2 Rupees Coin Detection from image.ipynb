```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('sl coins.jpg') #load image
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #convert image to gray format
plt.imshow(img2, cmap='gray')
```

```python
ret,img3 = cv2.threshold(img2,240,255,cv2.THRESH_BINARY) #add threshold to seaparate image into binary format, 
                                                         #if pixel value is greater than 240,convert to white and else black
plt.imshow(img3, cmap='gray')
```

```python
#find contours and hierarchy according to TREE retreval method and SIMPLE chain approximation method
contours, hierarchy = cv2.findContours(img3, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

print(len(contours)) #How many contours have
```

```python
#Assume we need to identify 2 coin. So as an specific feature, we are considering its area. 

for i in contours:
    area = cv2.contourArea(i)
    print(area)
```

```python
#We can identify that 2 coin area should be largest area
#it can be 18876
for i in contours:
    area = cv2.contourArea(i)
    
    if(area>18000 and area<20000):
        cv2.drawContours(img, [i], -1, (0,255,255), 2)
        x,y,w,h = cv2.boundingRect(i)
        print(x,y,w,h)
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)

cv2.imshow('img',img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
